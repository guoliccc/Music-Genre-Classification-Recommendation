```
from pydub import AudioSegment
import os
three_seconds = 3 * 1000# 单位：ms
```

```
for g in genres:
  print(g)
  for filename in os.listdir(f'/content/drive/My Drive/Colab Notebooks/5001/Data/genres_original/{g}/'):
    songname = f'/content/drive/My Drive/Colab Notebooks/5001/Data/genres_original/{g}/{filename}'
    sound = AudioSegment.from_wav(songname)

    for i in range(10):
      s = sound[three_seconds*i:three_seconds*(i+1)]
      names = filename.split('.')
      s.export(f"/content/drive/My Drive/Colab Notebooks/5001/Data/cut_file/{g}/{names[0]}.{names[1]}.{i+1}.wav", format="wav")
```
