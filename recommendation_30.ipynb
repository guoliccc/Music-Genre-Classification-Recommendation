```python
data = pd.read_csv("./data/features_30_sec.csv")
# data = data.iloc[0:, 2:]
data = data.drop(columns='length',axis=1)
print(data.shape)
data.head(10)
```

```python
data_3 = pd.read_csv("./data/features_3_sec.csv")
# data = data.iloc[0:, 2:]
data_3 = data_3.drop(columns='length',axis=1)
print(data_3.shape)
data_3.head(10)
```

# one KD tree

```python
tdata = data.drop(columns='label',axis=1)
tdata = tdata.drop(columns='filename',axis=1)
X = np.array(tdata.values.tolist())
sx = preprocessing.StandardScaler().fit(X)
ssx = sx.transform(X)
tree = KDTree(ssx, leaf_size=50)
```

```python
q = 345
qlabel = data.iloc[q]['label']
dist, ind = tree.query([ssx[q]], k=6)
print(dist)
print(ind)
data.iloc[ind[0]]
```

```python
#query 3s piece
tdata_3 = data_3.drop(columns='label',axis=1)
tdata_3 = tdata_3.drop(columns='filename',axis=1)
X_3 = np.array(tdata_3.values.tolist())
sx_3 = preprocessing.StandardScaler().fit(X_3)
ssx_3 = sx.transform(X_3)
q = 234
qlabel = data_3.iloc[q]['label']
dist, ind = tree.query([ssx_3[q]], k=6)
print(dist)
print(ind)
data.iloc[ind[0]]
```

```python
sx = preprocessing.StandardScaler().fit(X)
ssx = sx.transform(X)
tree = KDTree(ssx, leaf_size=50)
acc = 0
rnum = 5
for q in range(1000):
    qlabel = data.iloc[q]['label']
    dist, ind = tree.query([ssx[q]], k=rnum+1)
    matched = data.iloc[ind[0]].where(data.label==qlabel).dropna(axis=0, how='all').shape[0]
    acc = acc + matched -1
    if q%50==0:
        print(str(q) + " done")
acc /= (1000*rnum)
print(acc)
```
